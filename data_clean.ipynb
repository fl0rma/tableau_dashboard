```python
initial_data = initial_data.drop(['Unnamed: 0'], axis=1)
initial_data = initial_data.rename(columns={'Text': 'short_description', 'ID': 'gameid', 'Name':"name_2"})
new_order = ['gameid', 'name_2', 'short_description']
initial_data = initial_data[new_order]
```

```python
merged_data = pd.merge(retreived_data, initial_data, on='gameid')
```

```python
merged_data['name'] = merged_data['name'].fillna(merged_data['name_2'])
```

```python
new_categories = pd.read_csv('./new_categories.csv')
```

```python
board_games_data = pd.merge(merged_data, new_categories, on='gameid')
```

```python
# Function to assign complexity level
def assign_level(value):
    if value < 2:
        return 'Light'
    elif value < 3:
        return 'Medium Light'
    elif value < 4:
        return 'Medium'
    elif value < 5:
        return 'Medium Heavy'
    else:
        return 'Heavy'

# Assign categories to the Complexity column
board_games_data['level'] = board_games_data['complexity'].apply(assign_level)
```

```python
# Function to assign publish era
def assign_era(value):
    if value < 1960:
        return 'Before 1960'
    elif value < 1970:
        return '1960s'
    elif value < 1980:
        return '1970s'
    elif value < 1990:
        return '1980s'
    elif value < 2000:
        return '1990s'
    elif value < 2010:
        return '2000s'
    elif value < 2020:
        return '2010s'    
    elif value < 2030:
        return '2020s'    
    
# Assign categories to the Complexity column
board_games_data['era'] = board_games_data['year_published'].apply(assign_era)
```

```python
board_games_data['yes_sum_4+_players'] = board_games_data['yes_4+_player'] + board_games_data['yes_5_player'] + board_games_data['yes_6_player'] + board_games_data['yes_6+_player']
board_games_data['no_sum_4+_players'] = board_games_data['no_4+_player'] + board_games_data['no_5_player'] + board_games_data['no_6_player'] + board_games_data['no_6+_player']
board_games_data['best_sum_4+_players'] = board_games_data['best_4+_player'] + board_games_data['best_5_player'] + board_games_data['best_6_player'] + board_games_data['best_6+_player']
```
